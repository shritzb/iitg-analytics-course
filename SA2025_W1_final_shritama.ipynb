### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
cars_new = cars.copy()
group = cars_new.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=group, x='model_year', y='mpg', hue='origin', marker='o')

plt.title ('Average MPG over years by origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title = 'Origin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here

#Identify car names that appear in more than one model_year
consistent_car = cars.copy()
car_year_counts = consistent_car.groupby('name')['model_year'].nunique()
multi_year_names = car_year_counts[car_year_counts > 1].index

#For each such name, compute the standard deviation of mpg across years.
multi_year_cars = consistent_car[consistent_car['name'].isin(multi_year_names)]
variation_stats = multi_year_cars.groupby('name')['mpg'].agg(['std', 'count', 'mean']).reset_index()

#Car(s) with the lowest variation in mpg, with at least 2 appearances and std(mpg) < 1.0.
consistent_models = variation_stats[(variation_stats['std'] < 1.0) & (variation_stats['count'] >= 2)]
sorted_models = consistent_models.sort_values(by=['std', 'count'], ascending= [False, False])

#Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.
min_std = sorted_models['std'].min()
most_consistent = sorted_models[sorted_models['std']  == min_std]

final_output = most_consistent[['name', 'count', 'mean']]
final_output.columns = ['Model Name', 'Appearances', 'Average MPG']
car_named = final_output['Model Name']
print(f"The most consistent model is {car_named}")
print(final_output)
```
